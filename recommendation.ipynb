# Simple Recommendation System
## Created for CodSoft AI Internship

```python
movies = {
    "action": ["Avengers", "Mad Max", "John Wick"],
    "comedy": ["3 Idiots", "PK", "Hera pheri"],
    "romance": ["Kuch kuch hota hai", "Dil to pagal hai", "Aashiqui 2"],
    "horror": ["Annabelle", "Bhediya", "Tumbbad"],
    "motivational": ["Dangal", "Super 30", "Lakshya"]
}

print("Welcome to Movie Recommendation System ðŸŽ¬")
print("Available genres: action, comedy, romantic, horror, motivational")

user_choice = input("Enter your favorite genre: ").lower()

if user_choice in movies:
    print("Recooended movies for you:")
    for movie in movies[user_choice]:
        print("_", movie)
else:
    print("Sorry, genre not found. please choose from available genres.")
        
```
